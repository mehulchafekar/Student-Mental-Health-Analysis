# Business Objective

The primary business objective of this project is as follows:

1. **Improve Student Support Services:** Provide data-session recommendations to improve counseling services, support programs
    and stress management resources in educational institutions.
         
      - **Example**: If we find that many students are stressed because they don’t get enough sleep, the school can start a               program to educate students about healthy sleep habits and maybe even adjust the school schedule to allow more sleep.


2. **Increase Awareness:** Raise awareness among stakeholders (students, parents, educators) about the factors contributing to
     student stress and the importance of mental health.
     
     - **Example**: If the data shows that financial stress is a big problem, the school can hold workshops for students and              parents about managing finances and finding scholarships.


3. **Create Better Programs and Policies:** Creation of policies and programs that address the root causes of student stress 
     and promote a healthier academic environment.
   
     - **Example**: If a lot of students are feeling stressed about their studies, the school can create a mentorship program            where older students help younger ones with their studies.


4. **Use Resources Wisely:** Allocate resources more effectively by identifying the most critical areas needing
     help and support.

     - **Example**: If the data shows that counseling services are most needed by students in their final year, the school can            hire more counselors to specifically help these students during exam times.


5.  **Help Students Succeed:** By reducing stress, we can help students do better in school and stay in school.

     - **Example**: If we find that high stress levels are linked to lower grades, the school can create stress-relief programs,          like yoga or meditation classes, to help students manage stress and improve their academic performance.

5.**Skewness:** Skewness helps us understand the shape of our data distribution. Most columns show skewness values close to zero, indicating the distributions are relatively symmetrical.

let's go through the histograms explain above and understand how they help us check for skewness:

1. **Student_ID**:
   - **What it shows**: The Student_ID histogram shows a uniform distribution, meaning student IDs are evenly spread out without skewness.
   - **Skewness**: Zero skewness, which is expected because IDs are unique identifiers.

2. **Age**:
   - **What it shows**: Age peaks around certain points (e.g., 20, 22, 24), meaning most students are around these ages.
   - **Skewness**: Slight positive skew, as there are more younger students and fewer older ones.

3. **Year_of_Study**:
   - **What it shows**: Peaks at certain years, with more students in the 2nd year.
   - **Skewness**: Slight positive skew, indicating more students are in the earlier years.

4. **Sleep_Hours**:
   - **What it shows**: Sleep hours are fairly even, with most students sleeping between 4 to 8 hours.
   - **Skewness**: Nearly zero skewness, indicating a balanced distribution of sleep hours.

5. **Academic_Stress_Level**:
   - **What it shows**: Academic stress levels have peaks at multiple points.
   - **Skewness**: Slight negative skew, indicating most students have lower to moderate stress levels.

6. **Financial_Stress_Level**:
   - **What it shows**: Similar to academic stress, with peaks at certain points.
   - **Skewness**: Slight positive skew, indicating most students have low to moderate financial stress.

7. **Family_Stress_Level**:
   - **What it shows**: Even distribution across stress levels.
   - **Skewness**: Nearly zero skewness, indicating a balanced stress level from family issues.

8. **Emotional_Stress_Level**:
   - **What it shows**: Peaks at multiple points similar to other stress levels.
   - **Skewness**: Slight positive skew, indicating most students have low to moderate emotional stress.

9. **Total_Stress_Level**:
   - **What it shows**: More students have lower total stress levels, with fewer students having high total stress levels.
   - **Skewness**: Positive skew, showing a longer tail on the right side, indicating a few students have very high stress levels.
   
10. **Counseling_Need**
    - **What it shows**: This histogram has two distinct bars at 0 and 1, indicating a binary distribution.
    - **Skewness**: No skewness. The distribution is perfectly symmetrical because the values are either 0 (no need for counseling) or 1 (need for counseling). 

11. **CGPA**
    - **What it shows**: The CGPA values are distributed between 2.0 and 4.0, with a slight peak around 3.0 to 3.5.
    - **Skewness**: Slight negative skewness (left-skewed). This means most students have higher CGPAs (closer to 4.0), with fewer students having lower CGPAs.

12. **Parental_Income**
    - **What it shows**: Parental income ranges from 20,000 to 60,000, with multiple peaks indicating several common income levels.
    - **Skewness**: The distribution is relatively uniform with no clear skewness. It appears to be multimodal (having multiple peaks).

### Summary of Skewness in Simple Terms

- **Positive Skewness (Right-Skewed)**: More data points on the left, with a few stretching out to the right.
  - **Example**: Total_Stress_Level

- **Negative Skewness (Left-Skewed)**: More data points on the right, with a few stretching out to the left.
  - **Example**: Academic_Stress_Level

- **Zero Skewness (Symmetrical)**: Even distribution on both sides.
  - **Example**: Sleep_Hours, Family_Stress_Level

### 12. Find the Average Total Stress by year of Study?
- Line chart: Counseling Need?

```python
# Group by Year of Study and calculate the mean Total Stress Level
yearly_stress = df.groupby('Year_of_Study')['Total_Stress_Level'].mean().reset_index()

# Line Chart for Total Stress Levels by Year of Study
plt.figure(figsize=[6, 5])
sns.lineplot(x='Year_of_Study', y='Total_Stress_Level',color = 'g', data=yearly_stress,linestyle = 'dashed',marker= 'o',markersize = 8)
plt.title('Average Total Stress Level by Year of Study')
plt.xlabel('Year of Study')
plt.ylabel('Average Total Stress Level')
plt.grid()
plt.show()
```

### 14. Find relationships between four variables: CGPA, Total Stress Level, Sleep Hours, and Parental Income
- HeatMap: CGPA, Total Stress Level, Sleep Hours, and Parental Income

```python
plt.figure(figsize=(5,4))
correlation = df[['CGPA', 'Total_Stress_Level', 'Sleep_Hours', 'Parental_Income']].corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

- Here, Correlations are statistical measures that describe how two variables move in relation to each other. Hence here correlations helps to identify relationships between variables

### 17. Find the CGPA Distribution using KDE chart

- Here, in Kernel Density Estimate (KDE),

```python
sns.kdeplot(df['CGPA'], fill=True, color='purple')
plt.title('CGPA Distribution')
plt.xlabel('CGPA')
plt.show()
```

The peak of the KDE plot indicates the most common CGPA values. In this chart, the peak suggests that a particular range of CGPA values is more frequent among students.

The spread of the plot shows the variability in CGPA values. A wider spread indicates more variability, while a narrower spread indicates less variability.
